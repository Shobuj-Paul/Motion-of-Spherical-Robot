```python
from math import pi, sin, cos
import matplotlib.pyplot as plt
import numpy as np
```

```python
m = 1
l = 1
g = 9.81
h = 0.001
max_iterations = 10000
iterations = np.linspace(0,max_iterations,max_iterations)
E_analytic = 9.81*np.ones((max_iterations,), dtype=int)
```

### Euler Explicit Method

```python
q_0 = pi/2 #Initial Position
p_0 = 0 #Initial momentum
q = np.array([q_0])
p = np.array([p_0])
E = np.array([(p_0**2)/(2*m) - m*g*l*q_0])
for n in range(1,max_iterations):
    q_1 = q_0 + (h/(m*(l**2)))*p_0
    p_1 = p_0 - m*g*l*h*q_0
    E_1 = (p_1**2)/(2*m) - m*g*l*q_1
    q = np.append(q,q_1)
    p = np.append(p,p_1)
    E = np.append(E,E_1)
    q_0 = q_1
    p_0 = p_1
q_explicit = q
p_explicit = p
E_explicit = E
```

### Euler TV Method

```python
q_0 = pi/2 #Initial Position
p_0 = 0 #Initial momentum
q = np.array([q_0])
p = np.array([p_0])
E = np.array([(p_0**2)/(2*m) - m*g*l*q_0])
for n in range(1,max_iterations):
    q_1 = q_0 + (h/(m*(l**2)))*p_0
    p_1 = p_0 - m*g*l*h*q_1
    E_1 = (p_1**2)/(2*m) - m*g*l*q_1
    q = np.append(q,q_1)
    p = np.append(p,p_1)
    E = np.append(E,E_1)
    q_0 = q_1
    p_0 = p_1
q_tv = q
p_tv = p
E_tv = E
```

### Euler TV Method

```python
q_0 = pi/2 #Initial Position
p_0 = 0 #Initial momentum
q = np.array([q_0])
p = np.array([p_0])
E = np.array([(p_0**2)/(2*m) - m*g*l*q_0])
for n in range(1,max_iterations):
    p_1 = p_0 - m*g*l*h*q_0
    q_1 = q_0 + (h/(m*(l**2)))*p_1
    E_1 = (p_1**2)/(2*m) - m*g*l*q_1
    q = np.append(q,q_1)
    p = np.append(p,p_1)
    E = np.append(E,E_1)
    q_0 = q_1
    p_0 = p_1
q_vt = q
p_vt = p
E_vt = E
```

### Stormer Verlet Methods

```python
q_0 = pi/2 #Initial Position
p_0 = 0 #Initial momentum
q = np.array([q_0])
p = np.array([p_0])
E = np.array([(p_0**2)/(2*m) - m*g*l*q_0])
for n in range(1,max_iterations):
    p_half = p_0 - m*g*l*h*0.5*q_0
    q_1 = q_0 + (h/(m*(l**2)))*p_half
    p_1 = p_half - m*g*l*h*0.5*q_1
    E_1 = (p_1**2)/(2*m) - m*g*l*q_1
    q = np.append(q,q_1)
    p = np.append(p,p_1)
    E = np.append(E,E_1)
    q_0 = q_1
    p_0 = p_1
q_sv = q
p_sv = p
E_sv = E
```

```python
plt.plot(q_vt,p_vt,label="Euler Symplectic VT Method")
plt.plot(q_tv,p_tv,label="Euler Symplectic TV Methpd")
plt.plot(q_explicit,p_explicit,label="Euler Explicit Method")
plt.plot(q_sv,p_sv,label="Stormer-Verlet Method")
plt.xlabel("Angle")
plt.ylabel("Momentum")
plt.legend()
plt.show()
plt.plot(iterations,E_vt,label="Energy of Euler VT Method")
plt.plot(iterations,E_tv,label="Energy of Euler TV Method")
plt.plot(iterations,E_explicit,label="Energy of Euler Explicit Method")
plt.plot(iterations,E_sv,label="Energy of Stormer Verlet Method")
plt.plot(iterations,E_analytic,label = "Energy of Analytic Solution")
plt.xlabel("Iterations")
plt.ylabel("Energy")
plt.legend()
plt.show()
```
