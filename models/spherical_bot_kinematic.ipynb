```python
from math import pi, sin, cos
import matplotlib.pyplot as plt
import numpy as np
```

```python
m = 0.1
R = 1
I = 0.14
K = 1/(I + m*(R**2))
h = 0.01
p_alpha_0 = 0
p_theta_0 = 1.0
p_phi_0 = pi/2
alpha_0 = pi/2
theta_0 = pi
phi_0 = 0
x_0 = 0
y_0 = 0
```

```python
p_theta = np.array([p_theta_0])
theta = np.array([theta_0])
alpha = np.array([alpha_0])
phi = np.array([phi_0])
x = np.array([x_0])
y = np.array([y_0])
for n in range(1,10000):
    p_theta_1 = p_theta_0 + h*K*((p_alpha_0**2 + p_phi_0**2)*(cos(theta_0)/sin(theta_0)**3) - p_alpha_0*p_phi_0*((1+cos(theta_0)**2)/sin(theta_0)**3))
    theta_1 = theta_0 + h*K*p_theta_1
    alpha_1 = alpha_0 + h*K*(p_alpha_0*(m*R**2/I + 1/sin(theta_0)**2) - p_phi_0*cos(theta_0)/sin(theta_0)**2)
    phi_1 = phi_0 + h*K*((p_phi_0 - p_alpha_0*cos(theta_0))/sin(theta_0)**2)
    x_1 = x_0 + h*K*R*(p_theta_1*cos(alpha_1) + (p_phi_0 - p_alpha_0*cos(theta_1))*sin(alpha_1)/sin(theta_1))
    y_1 = y_0 + h*K*R*(p_theta_1*sin(alpha_1) - (p_phi_0 - p_alpha_0*cos(theta_1))*cos(alpha_1)/sin(theta_1))
    p_theta = np.append(p_theta, p_theta_1)
    theta = np.append(theta, theta_1)
    alpha = np.append(alpha, alpha_1)
    phi = np.append(phi, phi_1)
    x = np.append(x, x_1)
    y = np.append(y, y_1)
    p_theta_0 = p_theta_1
    alpha_0 = alpha_1
    phi_0 = phi_1
    theta_0 = theta_1
    x_0 = x_1
    y_0 = y_1
plt.plot(x,y,label="Path")
plt.legend()
plt.show()
```
