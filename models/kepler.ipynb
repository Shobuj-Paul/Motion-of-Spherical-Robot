```python
from math import pi, sin, cos, pow
import matplotlib.pyplot as plt
import numpy as np
```

```python
h = 0.01
G = 6.67408*pow(10,-11)
m = 1.0
M = 1.0
alpha = G*M
max_iterations = 100
iterations = np.linspace(0,max_iterations,max_iterations)
```

### Euler TV Method

```python
r_0 = 0.8
theta_0 = 0
p_r_0 = 0
p_theta_0 = 0.75
r = np.array([r_0])
theta = np.array([theta_0])
p_r = np.array([p_r_0])
E = np.array([(p_r_0**2 + (p_theta_0**2)/(r_0**2))/(2*m) - alpha*m/r_0])
for n in range(1,max_iterations):
    p_r_1 = p_r_0 + h*((p_theta_0**2)/(m*(r_0**3)) + alpha*m/(r_0**2))
    r_1 = r_0 + h*p_r_1/m
    theta_1 = theta_0 + h*(p_theta_0/r_0**2)
    E_1 = (p_r_1**2 + (p_theta_0**2)/(r_1**2))/(2*m) - alpha*m/r_1
    p_r = np.append(p_r, p_r_1)
    r = np.append(r, r_1)
    theta = np.append(theta, theta_1)
    E = np.append(E,E_1)
    r_0  = r_1
    theta_0 = theta_1
    p_r_0 = p_r_1
r_tv = r
theta_tv = theta
E_tv = E
```

### Euler Explicit Method

```python
r_0 = 0.8
theta_0 = 0
p_r_0 = 0
p_theta_0 = 0.75
r = np.array([r_0])
theta = np.array([theta_0])
p_r = np.array([p_r_0])
E = np.array([(p_r_0**2 + (p_theta_0**2)/(r_0**2))/(2*m) - alpha*m/r_0])
for n in range(1,max_iterations):
    p_r_1 = p_r_0 + h*((p_theta_0**2)/(m*(r_0**3)) + alpha*m/(r_0**2))
    r_1 = r_0 + h*p_r_0/m
    theta_1 = theta_0 + h*(p_theta_0/r_0**2)
    E_1 = (p_r_1**2 + (p_theta_0**2)/(r_1**2))/(2*m) - alpha*m/r_1
    p_r = np.append(p_r, p_r_1)
    r = np.append(r, r_1)
    theta = np.append(theta, theta_1)
    E = np.append(E,E_1)
    r_0  = r_1
    theta_0 = theta_1
    p_r_0 = p_r_1
r_explicit = r
theta_explicit = theta
E_explicit = E
```

```python
plt.plot(r_tv,theta_tv,label="Euler Symplectic TV Method")
plt.plot(r_explicit,theta_explicit,label="Euler Explicit Method")
plt.legend()
plt.show()
plt.plot(iterations,E_tv,label="Euler Symplectic TV Method")
plt.plot(iterations,E_explicit,label="Euler Explicit Method")
plt.legend()
plt.show()
```
