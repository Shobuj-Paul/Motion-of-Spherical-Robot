```python
from math import pi, sin, cos, pow
import matplotlib.pyplot as plt
import numpy as np
```

```python
h = 0.01
G = 6.67408*pow(10,-11)
m = 1.0
M = 1.0
alpha = G*M
```

```python
r_0 = 10
theta_0 = 0
p_r_0 = 0
p_theta_0 = 5
r = np.array([r_0])
theta = np.array([theta_0])
p_r = np.array([p_r_0])
x = np.array([])
y = np.array([])
for n in range(1,100000):
    p_r_1 = p_r_0 + h*((p_theta_0**2)/(m*(r_0**3)) + alpha*m/(r_0**2))
    r_1 = r_0 + h*p_r_1/m
    theta_1 = theta_0 + h*(p_theta_0/r_0**2)
    x_1 = r_1*cos(theta_1)
    y_1 = r_1*sin(theta_1)
    p_r = np.append(p_r, p_r_1)
    r = np.append(r, r_1)
    theta = np.append(theta, theta_1)
    x = np.append(x, x_1)
    y = np.append(y, y_1)
    r_0  = r_1
    theta_0 = theta_1
    p_r_0 = p_r_1
plt.plot(r,theta,label="Trajectory")
plt.show()
```
