```python
from math import pi, sin
import matplotlib.pyplot as plt
import numpy as np
```

```python
m = 1
l = 1
g = 9.81
h = 0.001
```

```python
q_0 = pi/2 #Initial Position
p_0 = 0 #Initial momentum
q = np.array([q_0])
p = np.array([p_0])
for n in range(1,2500):
    q_1 = q_0 + (h/(m*(l**2)))*p_0
    p_1 = p_0 - m*g*l*h*sin(q_1)
    q = np.append(q,q_1)
    p = np.append(p,p_1)
    q_0 = q_1
    p_0 = p_1
plt.plot(q,p,label="VT")
plt.legend()
plt.show()
```

```python
q_0 = pi - 0.01 #Initial Position
p_0 = 0 #Initial momentum
q = np.array([q_0])
p = np.array([p_0])
for n in range(1,8000):
    p_1 = p_0 - m*g*l*h*sin(q_0)
    q_1 = q_0 + (h/(m*(l**2)))*p_1
    q = np.append(q,q_1)
    p = np.append(p,p_1)
    q_0 = q_1
    p_0 = p_1
plt.plot(q,p,label="TV")
plt.legend()
plt.show()
```
